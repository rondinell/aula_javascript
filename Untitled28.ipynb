```python
ativo = "HBOR3.SA"
```

```python
dados_ativo = yf.download(ativo)

dados_ativo
```

```python
dados_ativo['retornos'] = dados_ativo['Close'].pct_change()
```

```python
dados_ativo['retornos_postivos'] = dados_ativo['retornos'].apply(lambda x: x if x > 0 else 0)
dados_ativo['retornos_negativos'] = dados_ativo['retornos'].apply(lambda x: abs(x) if x < 0 else 0)
```

```python
dados_ativo['media_retornos_positivos'] = dados_ativo['retornos_postivos'].rolling(window=30).mean()
dados_ativo['media_retornos_negativos'] = dados_ativo['retornos_negativos'].rolling(window=30).mean()
```

```python
dados_ativo['RSI'] = (100 - 100 / (1 + dados_ativo['media_retornos_positivos'] / dados_ativo['media_retornos_negativos']))
```

```python
dados_ativo.loc[dados_ativo['RSI'] < 30, 'compra'] = 'sim'
dados_ativo.loc[dados_ativo['RSI'] >= 30, 'compra'] = 'nao' # correct the logic of the conditional

dados_ativo['compra'] = dados_ativo['compra'].fillna('nao') # fill NaN values with 'nao'
```

```python
datas_compra = []
datas_venda = []

for i in range(len(dados_ativo)):
    # if "sim" in dados_ativo['compra'].iloc[i]: #this line was causing the error
    if dados_ativo['compra'].iloc[i] == 'sim': # we check directly if the value is 'sim'

        datas_compra.append(dados_ativo.iloc[i+1].name)

datas_compra
```

```python
plt.figure(figsize = (15, 8))
plt.scatter(dados_ativo.loc[datas_compra].index, dados_ativo.loc[datas_compra]['Close'], marker = '^',
            c = 'g') # replace data_compra for datas_compra
plt.plot(dados_ativo['Close'], alpha = 0.7)
```
